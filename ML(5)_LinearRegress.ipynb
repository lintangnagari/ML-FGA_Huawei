## 5.2.1 PREPARING DATA

```python
# Import the required modules NumPy for calculation and Matplotlib for drawing.
import numpy as np
import matplotlib.pyplot as plt

#This code is used only for Jupyter Notebook.
%matplotlib inline

# Define data and convert the list into an array.
x =[3,21,22,34,54,34,55,67,89,99]
x = np.array(x)
y = [1,10,14,34,44,36,22,67,79,90]
y = np.array(y)

# Display the data through the scatter chart.
plt.scatter (x,y)
```

## 5.2.2 DEFINING RELATED FUNCTION

```python
# The basic linear regression model is wx+b. In this example, the model is ax+b as a two-dimensional space is used.
def model(a,b,x):
   return a*x+b

# The mean square error loss function is the most commonly used loss function in the linear regression model.
def loss_function(a,b,x,y):
   num=len(x)
   predict=model(a,b,x)
   return (0.5/num)*(np.square(predict-y)).sum()

# The optimization function mainly uses the partial derivatives to update a and b.
def optimize(a,b,x,y):
   num=len(x)
   predict=model(a,b,x)
   da = (1.0/num) * ((predict -y)*x).sum()
   db = (1.0/num) * ((predict -y).sum())
   a = a - Lr*da
   b = b - Lr*db
   return a, b

# Perform function iteration to return a and b
def iterate(a,b,x,y,times):
   for i in range(times):
      a,b = optimize(a,b,x,y)
   return a, b
```

## 5.2.3 STARTING ITERATION

```python
# Initialize and display parameters.
a = np.random.rand(1)
print(a)
b = np.random.rand(1)
Lr = 1e-4
a,b = iterate(a,b,x,y,1)
prediction=model(a,b,x)
loss = loss_function(a, b, x, y)
print(a,b,loss)
plt.scatter(x,y)
plt.plot(x,prediction)
```

## 5.3.1 Question 1
Must the loss value return to zero when the raw data is modified?

### asnwer
No, the loss value does not necessarily have to return to zero when the raw data is modified. In machine learning and optimization, the loss value represents the discrepancy between the predicted output and the actual target. It is used to quantify how well the model is performing on the given data.

When raw data is modified, the loss value might change, but it does not guarantee that it will become zero. The purpose of training a machine learning model is to minimize the loss function, but in practice, reaching a loss value of exactly zero is uncommon and can indicate overfitting to the training data. Overfitting means the model has memorized the training data rather than learning general patterns, and it may not perform well on new, unseen data.
